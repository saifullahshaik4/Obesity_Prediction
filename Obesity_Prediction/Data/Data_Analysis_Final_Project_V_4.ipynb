```python
# Block 2: Data Cleaning and Preprocessing

print("=" * 60)
print("DATA CLEANING AND PREPROCESSING")
print("=" * 60)

# Create a copy for cleaning
df_clean = df.copy()

print("\n1. MISSING VALUES ANALYSIS:")
missing_values = df_clean.isnull().sum()
missing_percent = (missing_values / len(df_clean)) * 100
missing_info = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_percent
}).sort_values('Missing Count', ascending=False)

print("Missing Values Summary:")
if missing_info['Missing Count'].sum() > 0:
    print(missing_info[missing_info['Missing Count'] > 0])

    # Handle missing values
    print("\nHandling missing values...")
    # For numerical columns, fill with median
    numerical_cols = df_clean.select_dtypes(include=[np.number]).columns
    for col in numerical_cols:
        if df_clean[col].isnull().sum() > 0:
            df_clean[col].fillna(df_clean[col].median(), inplace=True)
            print(f"  • {col}: filled {missing_values[col]} missing values with median")

    # For categorical columns, fill with mode
    categorical_cols = df_clean.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        if df_clean[col].isnull().sum() > 0:
            df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)
            print(f"  • {col}: filled {missing_values[col]} missing values with mode")
else:
    print("No missing values found")

print("\n2. DATA QUALITY CHECKS:")

# Check for unrealistic values
print("Checking for unrealistic values:")
quality_issues = []

# Age check
if 'Age' in df_clean.columns:
    age_issues = ((df_clean['Age'] < 10) | (df_clean['Age'] > 100)).sum()
    if age_issues > 0:
        print(f"  Age: {age_issues} unrealistic values found")
        quality_issues.append('Age')
    else:
        print(f"  Age: Values range from {df_clean['Age'].min():.1f} to {df_clean['Age'].max():.1f}")

# Height check
if 'Height' in df_clean.columns:
    height_issues = ((df_clean['Height'] < 1.0) | (df_clean['Height'] > 2.5)).sum()
    if height_issues > 0:
        print(f"  Height: {height_issues} unrealistic values found")
        quality_issues.append('Height')
    else:
        print(f"  Height: Values range from {df_clean['Height'].min():.2f} to {df_clean['Height'].max():.2f}m")

# Weight check
if 'Weight' in df_clean.columns:
    weight_issues = ((df_clean['Weight'] < 30) | (df_clean['Weight'] > 300)).sum()
    if weight_issues > 0:
        print(f"  Weight: {weight_issues} unrealistic values found")
        quality_issues.append('Weight')
    else:
        print(f"  Weight: Values range from {df_clean['Weight'].min():.1f} to {df_clean['Weight'].max():.1f}kg")

if not quality_issues:
    print("Key variables have realistic value ranges")

print("\n3. OUTLIER DETECTION:")

def detect_outliers_iqr(series):
    """Detect outliers using IQR method"""
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    return outliers, lower_bound, upper_bound

# Check outliers in key variables
key_variables = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

outlier_summary = {}
for var in key_variables:
    if var in df_clean.columns:
        outliers, lower, upper = detect_outliers_iqr(df_clean[var])
        outlier_summary[var] = {
            'count': len(outliers),
            'percentage': len(outliers) / len(df_clean) * 100,
            'range': f'[{lower:.2f}, {upper:.2f}]',
            'max_outlier': outliers.max() if len(outliers) > 0 else 'None'
        }

outlier_df = pd.DataFrame(outlier_summary).T
print("Outlier Summary (using IQR method):")
print(outlier_df)

print("\n4. FEATURE ENGINEERING:")

# Calculate BMI
if all(col in df_clean.columns for col in ['Height', 'Weight']):
    df_clean['BMI'] = df_clean['Weight'] / (df_clean['Height'] ** 2)
    print("Created: BMI (Body Mass Index)")

# Create BMI categories
df_clean['BMI_category'] = pd.cut(df_clean['BMI'],
                                 bins=[0, 18.5, 25, 30, float('inf')],
                                 labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
print("Created: BMI_category")

# Create age groups
df_clean['age_group'] = pd.cut(df_clean['Age'],
                              bins=[0, 25, 35, 45, 100],
                              labels=['Young (≤25)', 'Adult (26-35)', 'Middle-aged (36-45)', 'Senior (46+)'])
print("Created: age_group")

# Create physical activity level
if 'FAF' in df_clean.columns:
    df_clean['activity_level'] = pd.cut(df_clean['FAF'],
                                       bins=[0, 1, 2, 3, float('inf')],
                                       labels=['Sedentary (0-1)', 'Low (1-2)', 'Moderate (2-3)', 'High (3+)'])
    print("Created: activity_level")

# Create water consumption level
if 'CH2O' in df_clean.columns:
    df_clean['water_intake_level'] = pd.cut(df_clean['CH2O'],
                                           bins=[0, 1.5, 2.5, float('inf')],
                                           labels=['Low (<1.5L)', 'Adequate (1.5-2.5L)', 'High (>2.5L)'])
    print("Created: water_intake_level")

# Create technology usage category
if 'TUE' in df_clean.columns:
    df_clean['tech_usage_category'] = pd.cut(df_clean['TUE'],
                                            bins=[0, 1, 2, float('inf')],
                                            labels=['Low (0-1h)', 'Moderate (1-2h)', 'High (2h+)'])
    print("Created: tech_usage_category")

# Create healthy lifestyle score
if all(col in df_clean.columns for col in ['FCVC', 'FAF', 'CH2O']):
    # Normalize values to 0-1 scale and create composite score
    fcvc_norm = (df_clean['FCVC'] - df_clean['FCVC'].min()) / (df_clean['FCVC'].max() - df_clean['FCVC'].min())
    faf_norm = (df_clean['FAF'] - df_clean['FAF'].min()) / (df_clean['FAF'].max() - df_clean['FAF'].min())
    ch2o_norm = (df_clean['CH2O'] - df_clean['CH2O'].min()) / (df_clean['CH2O'].max() - df_clean['CH2O'].min())

    df_clean['healthy_lifestyle_score'] = (fcvc_norm + faf_norm + ch2o_norm) / 3 * 100
    print("Created: healthy_lifestyle_score")

print("\n5. CATEGORICAL VARIABLE ENCODING:")

# Create binary variables for key categorical features
categorical_mappings = {}

# Gender encoding
if 'Gender' in df_clean.columns:
    df_clean['Gender_Male'] = (df_clean['Gender'] == 'Male').astype(int)
    df_clean['Gender_Female'] = (df_clean['Gender'] == 'Female').astype(int)
    categorical_mappings['Gender'] = df_clean['Gender'].unique()
    print("Encoded: Gender")

# Family history encoding
if 'family_history_with_overweight' in df_clean.columns:
    df_clean['family_history_yes'] = (df_clean['family_history_with_overweight'] == 'yes').astype(int)
    categorical_mappings['family_history'] = df_clean['family_history_with_overweight'].unique()
    print("Encoded: family_history_with_overweight")

# FAVC encoding (frequent consumption of high caloric food)
if 'FAVC' in df_clean.columns:
    df_clean['FAVC_yes'] = (df_clean['FAVC'] == 'yes').astype(int)
    categorical_mappings['FAVC'] = df_clean['FAVC'].unique()
    print("Encoded: FAVC")

# Smoking encoding
if 'SMOKE' in df_clean.columns:
    df_clean['SMOKE_yes'] = (df_clean['SMOKE'] == 'yes').astype(int)
    categorical_mappings['SMOKE'] = df_clean['SMOKE'].unique()
    print("Encoded: SMOKE")

# SCC encoding (calories consumption monitoring)
if 'SCC' in df_clean.columns:
    df_clean['SCC_yes'] = (df_clean['SCC'] == 'yes').astype(int)
    categorical_mappings['SCC'] = df_clean['SCC'].unique()
    print("Encoded: SCC")

print("\n6. FINAL DATASET INFO:")
print(f"Final dataset shape: {df_clean.shape}")
print(f"New columns added: {set(df_clean.columns) - set(df.columns)}")

print("\n7. TARGET VARIABLE ANALYSIS:")
print("Obesity Level Distribution (cleaned data):")
obesity_distribution = df_clean['NObeyesdad'].value_counts().sort_index()
for category, count in obesity_distribution.items():
    percentage = (count / len(df_clean)) * 100
    print(f"  • {category}: {count} ({percentage:.1f}%)")

print("\n8. KEY STATISTICS AFTER CLEANING:")
print("\nPhysical Characteristics:")
physical_stats = df_clean[['Age', 'Height', 'Weight', 'BMI']].describe()
print(physical_stats.round(2))

print("\nLifestyle Variables:")
lifestyle_stats = df_clean[['FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']].describe()
print(lifestyle_stats.round(2))
```

```python
# Block 4: Predictive Modeling - Linear Regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
import warnings
warnings.filterwarnings('ignore')

print("=" * 60)
print("PREDICTIVE MODELING - LINEAR REGRESSION")
print("=" * 60)

print("\n1. FEATURE PREPARATION AND SELECTION")

# Prepare dataset for modeling
df_model = df_clean.copy()

# Encode categorical variables for modeling
le_obesity = LabelEncoder()
le_caec = LabelEncoder()
le_calc = LabelEncoder()
le_mtrans = LabelEncoder()

# Encode target variable (obesity levels) for regression
df_model['NObeyesdad_encoded'] = le_obesity.fit_transform(df_model['NObeyesdad'])
obesity_mapping = dict(zip(le_obesity.classes_, le_obesity.transform(le_obesity.classes_)))
print("Obesity Level Encoding:")
for category, code in obesity_mapping.items():
    print(f"  • {category}: {code}")

# Encode other categorical variables
df_model['CAEC_encoded'] = le_caec.fit_transform(df_model['CAEC'])
df_model['CALC_encoded'] = le_calc.fit_transform(df_model['CALC'])
df_model['MTRANS_encoded'] = le_mtrans.fit_transform(df_model['MTRANS'])

# Select features for modeling (LIFESTYLE FACTORS ONLY - NO HEIGHT/WEIGHT)
feature_columns = [
    # Lifestyle and dietary habits
    'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE',

    # Demographics and binary lifestyle factors
    'Age', 'Gender_Male', 'family_history_yes', 'FAVC_yes', 'SMOKE_yes', 'SCC_yes',
    'CAEC_encoded', 'CALC_encoded', 'MTRANS_encoded',

    # Engineered lifestyle features
    'healthy_lifestyle_score'
]

# Prepare features and targets (BMI and Weight only)
X = df_model[feature_columns]
y_bmi = df_model['BMI']
y_weight = df_model['Weight']

print(f"\nFeatures selected: {len(feature_columns)} variables")
print("Feature categories:")
print("  • Lifestyle & Dietary: 5")
print("  • Demographics & Lifestyle Binary: 6")
print("  • Categorical (encoded): 3")
print("  • Engineered Lifestyle Features: 1")
print("\nIMPORTANT: Height and Weight EXCLUDED to test lifestyle factors impact only")

print("\n2. MODEL 1: PREDICTING BMI")

# Split data for BMI prediction
X_train_bmi, X_test_bmi, y_train_bmi, y_test_bmi = train_test_split(
    X, y_bmi, test_size=0.25, random_state=42)

print(f"Training set size: {X_train_bmi.shape[0]}")
print(f"Test set size: {X_test_bmi.shape[0]}")

# Scale features
scaler_bmi = StandardScaler()
X_train_bmi_scaled = scaler_bmi.fit_transform(X_train_bmi)
X_test_bmi_scaled = scaler_bmi.transform(X_test_bmi)

# Train linear regression model
lr_bmi = LinearRegression()
lr_bmi.fit(X_train_bmi_scaled, y_train_bmi)

# Make predictions
y_pred_bmi_train = lr_bmi.predict(X_train_bmi_scaled)
y_pred_bmi_test = lr_bmi.predict(X_test_bmi_scaled)

# Calculate metrics
bmi_train_rmse = np.sqrt(mean_squared_error(y_train_bmi, y_pred_bmi_train))
bmi_test_rmse = np.sqrt(mean_squared_error(y_test_bmi, y_pred_bmi_test))
bmi_train_r2 = r2_score(y_train_bmi, y_pred_bmi_train)
bmi_test_r2 = r2_score(y_test_bmi, y_pred_bmi_test)
bmi_test_mae = mean_absolute_error(y_test_bmi, y_pred_bmi_test)

print(f"\nBMI Prediction Results:")
print(f"  Training RMSE: {bmi_train_rmse:.3f}")
print(f"  Test RMSE: {bmi_test_rmse:.3f}")
print(f"  Training R²: {bmi_train_r2:.3f}")
print(f"  Test R²: {bmi_test_r2:.3f}")
print(f"  Test MAE: {bmi_test_mae:.3f}")

# Feature importance analysis
feature_importance_bmi = pd.DataFrame({
    'Feature': feature_columns,
    'Coefficient': lr_bmi.coef_,
    'Abs_Coefficient': np.abs(lr_bmi.coef_)
}).sort_values('Abs_Coefficient', ascending=False)

print(f"\nTop 10 Most Important Features for BMI Prediction:")
top_10_bmi = feature_importance_bmi.head(10)
for _, row in top_10_bmi.iterrows():
    direction = "↑ increases" if row['Coefficient'] > 0 else "↓ decreases"
    print(f"  • {row['Feature']}: {direction} BMI (coef: {row['Coefficient']:.3f})")

print("\n3. MODEL 2: PREDICTING WEIGHT")

# Split data for weight prediction
X_train_weight, X_test_weight, y_train_weight, y_test_weight = train_test_split(
    X, y_weight, test_size=0.25, random_state=42)

# Scale features
scaler_weight = StandardScaler()
X_train_weight_scaled = scaler_weight.fit_transform(X_train_weight)
X_test_weight_scaled = scaler_weight.transform(X_test_weight)

# Train linear regression model
lr_weight = LinearRegression()
lr_weight.fit(X_train_weight_scaled, y_train_weight)

# Make predictions
y_pred_weight_train = lr_weight.predict(X_train_weight_scaled)
y_pred_weight_test = lr_weight.predict(X_test_weight_scaled)

# Calculate metrics
weight_train_rmse = np.sqrt(mean_squared_error(y_train_weight, y_pred_weight_train))
weight_test_rmse = np.sqrt(mean_squared_error(y_test_weight, y_pred_weight_test))
weight_train_r2 = r2_score(y_train_weight, y_pred_weight_train)
weight_test_r2 = r2_score(y_test_weight, y_pred_weight_test)
weight_test_mae = mean_absolute_error(y_test_weight, y_pred_weight_test)

print(f"\nWeight Prediction Results:")
print(f"  Training RMSE: {weight_train_rmse:.3f}")
print(f"  Test RMSE: {weight_test_rmse:.3f}")
print(f"  Training R²: {weight_train_r2:.3f}")
print(f"  Test R²: {weight_test_r2:.3f}")
print(f"  Test MAE: {weight_test_mae:.3f}")

# Feature importance for weight
feature_importance_weight = pd.DataFrame({
    'Feature': feature_columns,
    'Coefficient': lr_weight.coef_,
    'Abs_Coefficient': np.abs(lr_weight.coef_)
}).sort_values('Abs_Coefficient', ascending=False)

print(f"\nTop 10 Most Important Features for Weight Prediction:")
top_10_weight = feature_importance_weight.head(10)
for _, row in top_10_weight.iterrows():
    direction = "↑ increases" if row['Coefficient'] > 0 else "↓ decreases"
    print(f"  • {row['Feature']}: {direction} weight (coef: {row['Coefficient']:.3f})")

print("\n4. ENHANCED MODEL: RANDOM FOREST (COMPARISON)")

# Train Random Forest for comparison on BMI prediction
rf_bmi = RandomForestRegressor(n_estimators=100, random_state=42)
rf_bmi.fit(X_train_bmi, y_train_bmi)
y_pred_rf_bmi = rf_bmi.predict(X_test_bmi)

rf_bmi_rmse = np.sqrt(mean_squared_error(y_test_bmi, y_pred_rf_bmi))
rf_bmi_r2 = r2_score(y_test_bmi, y_pred_rf_bmi)

# Train Random Forest for weight prediction
rf_weight = RandomForestRegressor(n_estimators=100, random_state=42)
rf_weight.fit(X_train_weight, y_train_weight)
y_pred_rf_weight = rf_weight.predict(X_test_weight)

rf_weight_rmse = np.sqrt(mean_squared_error(y_test_weight, y_pred_rf_weight))
rf_weight_r2 = r2_score(y_test_weight, y_pred_rf_weight)

print(f"\nRandom Forest vs Linear Regression Comparison:")
print(f"BMI Prediction:")
print(f"  Linear Regression R²: {bmi_test_r2:.3f}")
print(f"  Random Forest R²: {rf_bmi_r2:.3f}")
print(f"  Improvement: {((rf_bmi_r2 - bmi_test_r2) / bmi_test_r2 * 100):.1f}%")

print(f"\nWeight Prediction:")
print(f"  Linear Regression R²: {weight_test_r2:.3f}")
print(f"  Random Forest R²: {rf_weight_r2:.3f}")
print(f"  Improvement: {((rf_weight_r2 - weight_test_r2) / weight_test_r2 * 100):.1f}%")

# Random Forest feature importance
rf_feature_importance = pd.DataFrame({
    'Feature': feature_columns,
    'Importance': rf_bmi.feature_importances_
}).sort_values('Importance', ascending=False)

print(f"\nRandom Forest Feature Importance (BMI prediction):")
for _, row in rf_feature_importance.head(10).iterrows():
    print(f"  • {row['Feature']}: {row['Importance']:.3f}")

print("\n5. MODEL COMPARISON SUMMARY")

model_comparison = pd.DataFrame({
    'Model': ['BMI Prediction (Lifestyle Only)', 'Weight Prediction (Lifestyle Only)'],
    'Test R²': [bmi_test_r2, weight_test_r2],
    'Test RMSE': [bmi_test_rmse, weight_test_rmse],
    'Test MAE': [bmi_test_mae, weight_test_mae],
    'Model Quality': [
        'Excellent' if bmi_test_r2 > 0.9 else 'Good' if bmi_test_r2 > 0.7 else 'Moderate' if bmi_test_r2 > 0.5 else 'Needs Improvement',
        'Excellent' if weight_test_r2 > 0.9 else 'Good' if weight_test_r2 > 0.7 else 'Moderate' if weight_test_r2 > 0.5 else 'Needs Improvement'
    ]
})

print("\nModel Performance Summary:")
print(model_comparison.round(3))

print("\n6. LIFESTYLE FACTORS IMPACT ANALYSIS")
print("\nLifestyle Features Impact on Obesity-Related Outcomes:")
lifestyle_features = ['FCVC', 'FAF', 'CH2O', 'TUE', 'FAVC_yes', 'healthy_lifestyle_score']

for model_name, importance_df in [('BMI', feature_importance_bmi),
                                 ('Weight', feature_importance_weight)]:
    print(f"\n{model_name} Model - Lifestyle Impact:")
    lifestyle_impact = importance_df[importance_df['Feature'].isin(lifestyle_features)].head(6)
    for _, row in lifestyle_impact.iterrows():
        impact = "NEGATIVE" if row['Coefficient'] < 0 else "POSITIVE"
        print(f"  • {row['Feature']}: {impact} impact (coef: {row['Coefficient']:.3f})")
```
